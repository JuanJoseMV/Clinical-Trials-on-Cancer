```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler # Escala los datos
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
def loadCSV(pathSamples, pathMatrix):
    df_loaded = pd.read_table(pathMatrix, sep=',')
    data = pd.read_table(pathSamples, sep=',')
    clin_trial_values = df_loaded.values
    
    Y = data['Eligible']
    Y = Y.astype(int)
    X = clin_trial_values[:, :]
    return X, Y
```

```python
X, Y = loadCSV("../../Dataset/10k_1Col_NoCarEsp_LSA.csv", "../../Tables/docsTopicsLSA1500.csv") #Cargar SCV
```

```python
scaler = MinMaxScaler() #Escala entre 0 y 1
Xt = scaler.fit_transform(X)
```

```python
# http://featureselection.asu.edu/tutorial.php
```

```python
clf = LDA(tol=0.5, store_covariance=True)
clf.fit(X, Y)
clf.decision_function(X)
```

```python
efes = np.zeros(X.shape[1])
feature_reduction_index = []

for j in range(X.shape[1]):
    f = 0
    for k in range(X.shape[1]):
        if j == k:
            continue
        f += ((np.mean(X[:,j]) - np.mean(X[:,k]))**2) / (np.var(X[:,j]) + np.var(X[:,k]))
        
    print(f)
    efes[j] = f
        
print('Count:', len(feature_reduction_index))
```

```python
efes = np.zeros(X.shape[1])

for j in range(X.shape[1]):
    # Fisher Discriminant ratio for each class (0,1)
    f0 = ((np.mean(X[Y == 0, j]) - np.mean(X[Y == 1, j]))**2) / (np.var(X[Y == 0, j]) + np.var(X[Y == 1, j]))
    f1 = ((np.mean(X[Y == 1, j]) - np.mean(X[Y == 0, j]))**2) / (np.var(X[Y == 1, j]) + np.var(X[Y == 0, j]))
    f = f0 + f1
    efes[j] = f
        
# Min-Max Scaling
efes -= efes.min() 
efes /= efes.ptp()

# Threshold
feature_reduction_index = [efes > 0.5]
        
print('Count:', len(feature_reduction_index))
```
