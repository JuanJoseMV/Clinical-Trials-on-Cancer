```python
def loadCSV(pathSamples, pathMatrix):
    df_loaded = pd.read_table(pathMatrix, sep=',')
    data = pd.read_table(pathSamples, sep=',')
    clin_trial_values = df_loaded.values
    
    Y = data['Eligible']
    Y = Y.astype(int)
    X = clin_trial_values[:, :]
    return X, Y
```

```python
def saveMatrix(df, name):
    path = '../../Tables/'
    path += name + '.h5'
    df.to_hdf(path, key='df', index=False)
    return 
```

```python
def sequential_forwdard_selection(X, Y, criterion, d, num_processors):
    if __name__ ==  '__main__': 
        p = Pool(processes = num_processors)
        features = []
        
        for j in range(d):
            params = []
            
            for i in range(X.shape[1]):
                if i not in features:
                    new_features = features + [i]
                    params.append((X, Y, new_features))

            if criterion == 'sd': # Statistical distance
                scores = p.map(AtomicFunctions.multi_run_wrapper_sd, params)
            elif cirterion == 'cd': # Class distance
                scores = p.map(AtomicFunctions.multi_run_wrapper_cd, params)
                
            features.append(np.argmax(np.array(scores)))
            
            if (len(features) % 10 == 0):
                print(len(features))

        return features
```

```python
def select_k_from_score(score, num_features, threshold=None):
    scaler = MinMaxScaler() #Escala entre 0 y 1
    Xt = scaler.fit_transform(X)
    title = score
    
    if score == 'chi2':
        skb = SelectKBest(chi2, k=num_features)
        title = 'χ$^2$'
    elif score == 'F-value':
        skb = SelectKBest(f_classif, k=num_features)
    elif score == 'Mutual information':
        skb = SelectKBest(mutual_info_classif, k=num_features)
        
    X_new = skb.fit_transform(Xt, Y)
    
    # Plot scores
    plt.figure(figsize=(10,10))
    plt.plot(range(0,len(skb.scores_),1),skb.scores_)
    axes = plt.gca()
    
    if threshold is not None:
        plt.axhline(y=threshold, color='r', linestyle='-')
        
    plt.ylabel(title)
    plt.xlabel('Feature index')
    plt.show()
    
    return X_new
```
